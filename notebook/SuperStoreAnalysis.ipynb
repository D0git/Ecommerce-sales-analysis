**Data Cleaning**

```python
# Total of nulls
df.isnull().sum()
```

A check for missing values revealed that all columns are complete except for the "Postal Code" column, which contains 11 missing entries. This indicates that the dataset is largely clean, with only a minor issue in one column that requires handling.

```python
# Handling null values in Postal Code column
df['Postal Code'] = df['Postal Code'].fillna(df['Postal Code'].mean())

# Checking another time
df.isnull().sum()
```

The missing values in the "Postal Code" column were handled using mean imputation to preserve the dataset's completeness without removing any records.

After applying this method, a verification step confirmed that all missing values were successfully resolved, resulting in a fully complete dataset ready for analysis.

```python
# Total of duplicates
df[df.duplicated()]
```

A check for duplicate records was performed, and no duplicate entries were found in the dataset.

This indicates that the data is consistent and does not require any further deduplication, ensuring the reliability of the analysis.

***Analysis***

```python
# Total Revenue
total_revenue = df['Sales'].sum()
print('Total revenue is : ',total_revenue)
```

# The total revenue generated is approximately $2.26 million, indicating a strong overall sales performance and a solid market presence.

```python
# Sales by Category
df.groupby('Category')['Sales'].sum().sort_values(ascending= False)
```

The Technology category is the top-performing category in terms of total sales, indicating strong customer demand for technology-related products and making it a key revenue driver for the business.

```python
# Sales by Sub-Category
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending= False)
```

At a more detailed level, sub-categories such as Phones and Chairs generate the highest sales, highlighting them as the primary contributors to revenue.

In contrast, sub-categories like Fasteners, Labels, and Envelopes show significantly lower sales, indicating potential underperformance or lower demand in these product areas.

```python
# Sales by Region
df.groupby('Region')['Sales'].sum()
```

The West region generates the highest sales, followed by the East region, indicating that these areas are the strongest markets for the business.

In contrast, the South region shows the lowest performance, suggesting potential opportunities for growth and market expansion.

```python
# Top Customers
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending = False).head(10)
```

A small group of customers contributes significantly to total revenue, with top customers such as Sean Miller and Tamara Chand generating the highest sales.

This indicates the presence of high-value customers who play a crucial role in driving overall business performance.

```python
# Yearly Trend
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['year'] = df['Order Date'].dt.year
df.groupby('year')['Sales'].sum()
```

Sales show an overall upward trend over the years, despite a slight decline in 2016.
Growth becomes significant from 2017 onwards, reaching the highest level in 2018, which indicates strong business expansion and improved performance.

The date column was parsed using day-first format to ensure correct datetime conversion.

```python
# Shipping Analysis

df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
df['delivery_days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['delivery_days'].mean()
```

The average delivery time is approximately 4 days, indicating an efficient shipping process.

However, further optimization could enhance customer satisfaction and provide a competitive advantage.

```python
# Sales by Segment
df.groupby('Segment')['Sales'].sum()
```

The Consumer segment generates the highest share of sales, significantly outperforming Corporate and Home Office segments.

This suggests that individual customers are the primary drivers of revenue for the business
